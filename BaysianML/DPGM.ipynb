# Dhirichlet Process Mixture Gauss Model
無限次元ガウス混合モデルの実装

実験用データ生成パイプライン

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#二次元正規分布（対角）
def generate(mu=[[-1,1],[1,1],[0,-1]],Sigma=[0.5,0.5,0.5],nk=[40,25,30],verb=True):
    X = np.concatenate([np.random.multivariate_normal(mu[k],Sigma[k]*np.eye(2),size=nk[k]) for k in range(len(mu))])
    y = np.concatenate([np.full(nk[k],k) for k in range(len(mu))])
    print(X.shape)
    
    if verb:
        plt.scatter(X[:,0],X[:,1],c=y)
        plt.show()
        
    rand_id = np.random.shuffle(np.arange(len(y)))
    return X[rand_id],y[rand_id]
```

```python
X,y = generate(mu=[[-2,3],[2,2],[0,-1]],Sigma=[0.1,0.2,0.3],nk=[40,25,30],verb=True)
```
