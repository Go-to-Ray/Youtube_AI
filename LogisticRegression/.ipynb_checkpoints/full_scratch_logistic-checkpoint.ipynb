**補足**
正則化項ですが$L_1$ノルム、$L_2$ノルムで効果が違っていて、ペナルティのかけ方が違います。詳しくは「$L_p$ノルム」や「正則化項」で検索するとわかります。
ここでは特に$L_1$ノルムとそれを制約に与えるメリットについて話します。
$L_1$ノルムは$\sum|\omega_i|$となっていてつまり、係数の絶対値が大きいものの影響を大きくできるということです。これによって、効果の弱い次元を削減できます。

以下のコードでわかりやすく感じてもらえると嬉しいです。
以下のコードはa,bが係数になっていてその組み合わせのメッシュを切って、係数a,bがあるときの時の損失関数を高等線として表しています。
正則化項を加えると、軸方向に尖ります。つまりある軸の要素を重要視しているといえます。（ということが伝わると嬉しいです）

```
n = 100
a = np.linspace(-1, 1, n)
b = np.linspace(-1, 1, n)
 
A, B = np.meshgrid(a, b)
Cost = np.sqrt(A**2 + B**2) + 1*(np.abs(A) + np.abs(B))
plt.contour(A, B, Cost)
plt.gca().set_aspect('equal')
plt.show()
```

````python
%matplotlib inline
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, interactive_output, Output
%matplotlib inline
````

````python
n = 200

a = np.linspace(-10, 10, n)
b = np.linspace(-10, 10, n)
x = np.linspace(-10, 10, n)
y = np.linspace(-10, 10, n)

gamma_step = 50
gamma_lst = np.linspace(0, 2, gamma_step)
min_lst = [[0]*2 for i in range(gamma_step)]

X, Y = np.meshgrid(x, y) 
A, B = np.meshgrid(a, b)
Cost = np.sqrt((X-8)**2 + (Y-2)**2)
C = (np.abs(A) + np.abs(B))

for i,gamma in enumerate(gamma_lst):
    loss = Cost + gamma*C
    A_min,B_min = np.where(loss == np.min(loss))
    min_lst[i][0] = b[B_min.astype(int)]
    min_lst[i][1] = a[A_min.astype(int)]
    
min_lst = np.array(min_lst)
"""
plt.contour(A, B, Cost)
plt.contour(A, B, C)
plt.plot(min_lst[:,0],min_lst[:,1])
plt.gca().set_aspect('equal')
plt.show()
"""

plt.contour(A, B, loss)
plt.plot(min_lst[:,0],min_lst[:,1])
plt.gca().set_aspect('equal')
plt.show()
````

````python
%matplotlib widget
%matplotlib inline
n = 50

a = np.linspace(-10, 10, n)
b = np.linspace(-10, 10, n)
x = np.linspace(-10, 10, n)
y = np.linspace(-10, 10, n)
gamma_step = 100
gamma_lst = np.linspace(0.2, 1.0, gamma_step)

X, Y = np.meshgrid(x, y) 
A, B = np.meshgrid(a, b)
Cost = np.sqrt((X-8)**2 + (Y-2)**2)
C = (np.abs(A) + np.abs(B))

loss = []
A_min = []
B_min = []
for i in range(gamma_step):
    loss.append(Cost + gamma_lst[i]*C)
    A_min.append(np.where(loss[i] == np.min(loss[i]))[0])
    B_min.append(np.where(loss[i] == np.min(loss[i]))[1])

@interact(step=(0,gamma_step-1,1))
def animation(step=0):
    plt.contour(A, B, loss[step])
    plt.scatter(b[B_min[step].astype(int)],a[A_min[step].astype(int)])
    plt.axhline(0, ls = "--", color = "c")
    plt.axvline(0, ls = "--", color = "c")
    plt.gca().set_aspect('equal')
    plt.savefig("L1_"+str(step)+".png")
    plt.show()
````
